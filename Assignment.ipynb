```python
import pandas as pd
from datetime import timedelta
```

```python
#def process_excel_file(file_path):
    # Load Excel file into a pandas DataFrame
df = pd.read_excel("Assignment_Timecard.xlsx")
```

```python
 # Convert the 'Time' and 'Time Out' columns to datetime format
df['Time'] = pd.to_datetime(df['Time'])
df['Time Out'] = pd.to_datetime(df['Time Out'])

# Sort the DataFrame by 'Employee Name' and 'Time'
df.sort_values(by=['Employee Name', 'Time'], inplace=True)
df['Timecard Hours (as Time)'] = pd.to_numeric(df['Timecard Hours (as Time)'], errors='coerce').astype(float)
```

```python
 # Iterate over the records
for index, row in df.iterrows():
    name = row['Employee Name']
    position_id = row['Position ID']
    time = row['Time']
    hours_worked = row['Timecard Hours (as Time)']

    # Check for more than 14 hours in a single shift
    if hours_worked > 14:
        print(f"{name} (Position ID: {position_id}) worked more than 14 hours.")

    # Print employees who worked 7 consecutive days
    consecutive_days = df[(df['Employee Name'] == name) & (df['Time'].between(time, time + timedelta(days=6)))]
    if len(consecutive_days) == 7:
        print(f"{name} (Position ID: {position_id}) worked 7 consecutive days.")

    # Check for less than 10 hours between shifts but greater than 1 hour
    next_shift = df[(df['Employee Name'] == name) & (df['Time'] > time)].head(1)
    if not next_shift.empty:
        hours_between_shifts = (next_shift['Time'].iloc[0] - time).total_seconds() / 3600
        if 1 < hours_between_shifts < 10:
            print(f"{name} (Position ID: {position_id}) has less than 10 hours between shifts.")
            
    print("\n")
```
